## Problem 1
1. Create the below pattern using nested for loop in Python.


*
**
***
****
*****
****
***
**
*

```python
n = 5

for i in range(n):
    print("*"*i)
for i in range(n,0,-1):
    print("*"*i)
```

## Problem 2

2. Write a Python program to reverse a word after accepting the input from the user.

Sample:

INPUT : INEURON
OUTPUT : NORUENI

```python
def rev_word(s):
    if len(s) == 0:
        return s
    else:
        return rev_word(s[1:]) + s[0]
    
s = input("Enter the word that needs to be reversed: ")

print(rev_word(s))
```
