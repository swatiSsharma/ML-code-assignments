# Problem 1 

Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200    both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
# First method

a = ""
for i in range(2000, 3201):
    if i%7 == 0 and i%5 != 0:
        if len(a) == 0:
            a = str(i)
        else:
            a = a + ", "+ str(i)
print(a)
        
```

```python
items = []

for i in range( 2000, 3201):
    if i%7 ==0 and i%5 != 0:
        items.append(i)

print(*items, sep = ", ") 
```

# Problem 2

Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

```python
fname = input("Enter the First name : ")
lname = input("Enter the Last name : ")
print(fname[::-1]+ " " + lname[::-1])
```

```python
def rev_name(input_str):
    s = ""
    for i in input_str:
        s =  i + s
    return s

fname = input("Enter the First name : ")
lname = input("Enter the Last name : ")

input_str = fname + " " + lname

print(rev_name(input_str))
```

# Problem 3

Write a Python program to find the volume of a sphere with diameter 12 cm. (Formula: V=4/3 * π * r 3)

```python
dia = 12
radius = dia/2

volume = round(4/3 * 3.14 * (radius ** 3),2)

superscript = str.maketrans("3", "³")

print("The volumne of sphere with diameter 12 is : {} ".format(volume) + "cm3".translate(superscript))
```
